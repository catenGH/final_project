```python
import tkinter as tk
import tkinter.ttk

root = tk.Tk()
root.title("Fridge to Food Base")

# main frame
mainframe = tk.Frame(root)
mainframe.pack()

# heading
heading = tk.Frame(mainframe, bg="mediumorchid3")
heading.grid(row=0, column=0, padx=10, pady=10)
heading.columnconfigure(1, weight=1)

# title in heading
title = tk.Label(heading, text="Fridge to Food", font=("Helvetica 24 bold", 20, "bold"), bg="mediumorchid3", fg="white", anchor="w")
title.grid(row=0, column=0, columnspan=2)

# input entry widget
ingredient = tk.StringVar()
ingredients_input = tk.Entry(heading, textvariable=ingredient, width=50)
ingredients_input.grid(row=0, column=3, columnspan=3, padx=(80, 0))

# add button
add_ingredient = tk.Button(heading, text="âœš")
add_ingredient.grid(row=0, column=6, padx=(10,0))

# search button
search = tk.Button(heading, text="ðŸ”Ž")
search.grid(row=0, column=7, padx=10)

# filter frame
filters = tk.Frame(mainframe, bg="gray", width=200)
filters.grid(row=1, column=0, columnspan=2, padx=10, pady=10, sticky="n")

# filter title
filter_title = tk.Label(filters, text="Filters")
filter_title.grid(row=2, column=0)

tkinter.ttk.Separator(mainframe, orient="vertical").grid(column=1, row=2, rowspan=7)

# cook times (less than 60 minutes is beginner, between 60 and 120 is intermediate, above 120 is expert)
cooktime = tk.Frame(filters, bg="gray")
cooktime.grid(row=2, column=0)

cooktime_title = tk.Label(cooktime, text="Cook Time (in minutes)", bg="gray", fg="white")
cooktime_title.pack()

beginner = tk.Checkbutton(cooktime, text="<60 (beginner)", bg="gray")
beginner.pack()

intermediate = tk.Checkbutton(cooktime, text=">=60 & =<120 (intermediate)", bg="gray")
intermediate.pack()

expert = tk.Checkbutton(cooktime, text=">120 (expert)", bg="gray")
expert.pack()

# cuisines
# african, asian, american, british, cajun, caribbean, chinese
# eastern european, european, french, german, greek, indian, irish, italian
# japanese, jewish, korean, latin american, mediterranean, mexican
# middle eastern, nordic, southern, spanish, thai, vietnamese
cuisines = tk.Frame(filters, bg="gray")
cuisines.grid(row=3, column=0)

cuisines_label = tk.Label(cuisines, text="Cuisines", bg="gray", fg="white")
cuisines_label.pack()

cuisine_list = ["African", "Asian", "American", "British", "Cajun", "Caribbean", "Chinese",
                "Eastern European", "European", "French", "German", "Greek", "Indian", "Irish", "Italian",
                "Japanese", "Jewish", "Korean", "Latin American", "Mediterranean", "Mexican",
                "Middle Eastern", "Nordic", "Southern", "Spanish", "Thai", "Vietnamese"]

for i in cuisine_list:
    cuisine = tk.Checkbutton(cuisines, text=i, bg="gray")
    cuisine.pack()

# recipe frame
recipes = tk.Frame(mainframe, bg="white")
recipes.grid(row=1, column=1, sticky="nsew", padx=10, pady=10)

# # recipes
# recipe_box = tk.Frame(recipes, bg="gray", bd=2, padx=10, pady=10)
# recipe_box.pack(padx=10, pady=10, fill="x")

# recipe_title = tk.Label(recipe_box, text="Example", bg="gray")
# recipe_title.pack(anchor="w")

# scrollbar
scroll = tk.Scrollbar(root, orient="horizontal")
scroll.pack(side="right")

root.mainloop()
```
