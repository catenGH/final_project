```python
def cooktime_filter(list):
    time = int(input("Pick a category: 1/2/3")) # replace this for the GUI with the checkboxes, maybe use if statements to see if checkbox is marked
    if time == 1:
        list = list.loc[list["Cook Time (in minutes)"]<60]
    if time == 2:
        list = list.loc[(list["Cook Time (in minutes)"] >= 60) & (list["Cook Time (in minutes)"] < 120)]
    if time == 3:
        list = list.loc[list["Cook Time (in minutes)"]>=120]
    return list

cooktime_filter(recipe)
```
