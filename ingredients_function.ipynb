```python
import requests
import json
```

```python
ingredients_api = requests.get("https://www.themealdb.com/api/json/v1/1/list.php?i=list")
ingredients_json = ingredients_api.json()
ingredients_dict = ingredients_json["meals"]
ingredients_dict
```

```python
ingredients_list = []

for ingredient in ingredients_dict:
    name = ingredient["strIngredient"]
    ingredients_list.append(name)

ingredients_list.sort()
print(ingredients_list)
print(len(ingredients_list))
# not sure if this is all of the ingredients in the thing but i'll save it
```

```python
ingredients_list_2 = []

for ingredient in ingredients_list:
    ingredient_name = ingredient.lower()
    ingredients_list_2.append(ingredient_name)

print(ingredients_list_2)
```

```python
# user input
main_ingredient = input("What is the main ingredient in your fridge?").lower()
if main_ingredient in ingredients_list_2:
    print("We have recipes for that!") # can replace later with filtering by recipe from spoonacular
else:
    print("We don't have this, pick something else!")
```

```python
# this is a function to input the user's main ingredient and get a list of dish names and pictures, can also tweak it to get recipes

def ingredient_input():
    main_ingredient = input("What is the main ingredient in your fridge?").lower()
    if main_ingredient in ingredients_list_2:
        main_ingredient = main_ingredient.replace(" ", "_")
        meal_api = requests.get(f"https://www.themealdb.com/api/json/v1/1/filter.php?i={main_ingredient}")
        meal_json = meal_api.json()
        results = {}
        for meal in meal_json["meals"]:
            meal_name = meal.get("strMeal")
            meal_ID = meal.get("idMeal")
            recipe_api = requests.get(f"https://www.themealdb.com/api/json/v1/1/lookup.php?i={meal_ID}")
            recipe_json = recipe_api.json()
            if recipe_json.get("meals"):
                recipe_ingredients = []
                recipe_pic = recipe_json["meals"][0]["strMealThumb"]
                recipe_instructions = recipe_json["meals"][0]["strInstructions"].replace("\r\n", " ")
                recipe = recipe_json["meals"][0]["strSource"]
                for key in recipe_json["meals"][0]:
                    if "strIngredient" in key:
                        ingredient_name = recipe_json["meals"][0][key]
                        if ingredient_name:
                            recipe_ingredients.append(ingredient_name)
                        else:
                            continue
                results[meal_name] = [recipe_pic, recipe_ingredients, recipe_instructions, recipe]
        print(results)
    else:
        print("We don't have this, pick something else!")

ingredient_input()
```
